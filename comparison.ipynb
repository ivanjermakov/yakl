# Efficiency comparison

```python
from matplotlib import pyplot as plt

import text as t
from efficiency import calculate
from layouts import *
from plot import bar_plot

layouts = [
    qwerty,
    dvorak,
    colemak
]

layoutsPenalties = list(map(lambda l: calculate(l, t.load_text(t.TEXT_PATH)), layouts))
```

```python
data = {
        "qwerty": [layoutsPenalties[0].distance],
        "dvorak": [layoutsPenalties[1].distance],
        "colemak": [layoutsPenalties[2].distance],
}

fig, ax = plt.subplots()
bar_plot(ax, data, total_width=.8, single_width=.9)
plt.xticks(range(1), ["Penalty for finger travel"])
plt.show()
```

```python
data = {
        "qwerty": [layoutsPenalties[0].finger],
        "dvorak": [layoutsPenalties[1].finger],
        "colemak": [layoutsPenalties[2].finger],
}

fig, ax = plt.subplots()
bar_plot(ax, data, total_width=.8, single_width=.9)
plt.xticks(range(1), ["Penalty for the same finger usage"])
plt.show()
```

```python
data = {
        "qwerty": [layoutsPenalties[0].keys],
        "dvorak": [layoutsPenalties[1].keys],
        "colemak": [layoutsPenalties[2].keys],
}

fig, ax = plt.subplots()
bar_plot(ax, data, total_width=.8, single_width=.9)
plt.xticks(range(1), ["Penalty for the same hand usage"])
plt.show()
```

```python
data = {
        "qwerty": [layoutsPenalties[0].hand],
        "dvorak": [layoutsPenalties[1].hand],
        "colemak": [layoutsPenalties[2].hand],
}

fig, ax = plt.subplots()
bar_plot(ax, data, total_width=.8, single_width=.9)
plt.xticks(range(1), ["Penalty for the non-dominant hand usage"])
plt.show()
```
