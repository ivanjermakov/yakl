# Efficiency

## Efficiency definition

Keyboard layout efficiency is a keyboard layout characteristic that measures comfortability and productivity (typing speed).
However, it does not consider not directly related aspects like common layout habits or mechanical layout peculiarities.

Efficiency is defined by the following aspects.

| Aspect name                          | Weight |
| :-                                   | -:     |
| finger travel distance               | .4     |
| consecutive keys on the same finger  | .3     |
| consecutive keys on the same hand    | .1     |
| dominant hand                        | .1     |
| pinky usage                          | .1     |

Each aspect has its weight, that is used to calculate efficiency.

## Efficiency testing

Efficiency test is represented by iterating over test text letter by letter.
On each letter performed calculation of *penalty* - a sum of efficiency aspects, listed above.
After iteration is complete, calculated average penalty of typed letters:

\begin{equation*}
avg = \frac{1}{n} \sum_{n = 1}^{k} p_n
\end{equation*}

Where $k$ - test text length, $p_n$ - typed letter penalty.

\begin{equation*}
p = \sum_a p_a * w_a
\end{equation*}

Where $p_a$ - aspect penalty, $w^a$ - aspect weight.

## Penalty

Each aspect has its own way of calculating penalty.

## Consecutive keys on the same finger

If current key is assigned to the same finger as the previous key, than this rule is activated.
Possible cases (examples are in _QWERTY_):

| Case             | Example    | Penalty |
| :--              | ---        | -:      |
| Same key         | `j` -> `j` | .2      |
| Adjacent key     | `j` -> `h` | .5      |
| Diagonal key     | `j` -> `u` | .6      |
| Over the row key | `m` -> `u` | 1       |

## Consecutive keys on the same hand

Activate rule is both current key and the previous one are assigned to the same hand.

## Dominant hand

Since [about 90% of population](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3927078/)
 are right handed, this rule is activated when key's assigned hand is left.

## Pinky usage

Since pinky finger has lower performance, rule is activated when key's assigned finger is pinky.
